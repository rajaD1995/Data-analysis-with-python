# Object oriented API

```python
# Object oriented API. ie figure and axis will be created under a variable.
# First create a grid of plots ie subplots.
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-');
```

---> we can create any size of graph and anywhere with the help of .add_axes([left, bottom, width, height]) function.

```python
# subplots without subplot() function and plot within plot without .subplot() function.
# By using .add_axes([left, bottom, width, height]) function.
# can create multiple axis within the fig or canvas by a .add_axes() function.
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig = plt.figure(figsize=(8, 6))  # Create figure

# Large main plot
ax1 = fig.add_axes([0.1, 0.3, 0.65, 0.6])  # [left, bottom, width, height]
#after creating axis plot the graph.
ax1.plot(x, y1, label="Sine Wave", color="blue")
# ax1.set_xlabel('x2')
# ax1.set_ylabel('y2')
ax1.set_title("Main Plot (Sine Wave)")
ax1.legend()

# Small subplot 1 (below main plot)
ax2 = fig.add_axes([0.1, 0.1, 0.3, 0.1])  
ax2.plot(x, y2, label="Cosine Wave", color="red")
ax2.set_title("Small Subplot 1")
ax2.legend()

# Small subplot 2 (beside subplot 1)
ax3 = fig.add_axes([0.45, 0.1, 0.3, 0.1])  
ax3.plot(x, y3, label="Tangent Wave", color="green")
ax3.set_title("Small Subplot 2")
ax3.legend()

# Inset inside main plot
ax_inset = fig.add_axes([0.34, 0.7, 0.15, 0.15])  
ax_inset.plot(x, y1**2, label="Sine Squared", color="purple")
ax_inset.set_title("Inset Plot")
ax_inset.legend(loc='best', bbox_to_anchor=(0.5, -0.7, 0.5, 0.5)) #legend argu for its placng,bboc-- eft, bottom, width, height
plt.tight_layout()  #so that title not ovelap. for equidistance among graphes.
plt.show()
```

```python
#subplots with loop and subplot() functions
fig, axes = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grid of subplots

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Flatten the 2D array of axes and loop through them
for i, ax in enumerate(axes.flatten()):
    ax.plot(x, np.sin(x + i), label=f"Shift {i}")
    ax.set_title(f"Plot {i+1}")
    ax.legend()

plt.tight_layout()  # Adjust layout to prevent overlap
# plt.show()
```

#streamlit project, here matplotlib Object-Oriented API

import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load titanic dataset
@st.cache_data
def load_data():
    data = pd.read_csv(r'C:\Users\USER\Documents\Python\NareshIT\21 feb 2025 eda streamlit\titanic dataset.csv')
    return data

data = load_data()
# Title and description
st.title('Exploratory Data Analysis of Titanic Dataset')
st.write('This is an EDA on the Titanic Dataset')
st.write('First few rows of the dataset: ')
st.dataframe(data.head())

# Data Cleaning Section
st.subheader('Missing Values')
missing_data = data.isnull().sum()
st.write(missing_data)

if st.checkbox('Fill missing Age with median'):
    data['Age'].fillna(data['Age'].median(), inplace=True)

if st.checkbox('Fill missing Embarked with mode'):
    data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

if st.checkbox('Drop duplicates'):
    data.drop_duplicates(inplace=True)

st.subheader('Cleaned Dataset')
st.dataframe(data.head())

# EDA Section
st.subheader('Statistical Summary of the Data')
st.write(data.describe())

# Age Distribution
st.subheader('Age Distribution')
fig, ax = plt.subplots()  #creatinf figue and axis
sns.histplot(data['Age'], kde=True, ax=ax)   #draw a graph in that figure with axis
ax.set_title('Age Distribution')
st.pyplot(fig)

# Gender Distribution
st.subheader('Gender Distribution')
fig, ax = plt.subplots()
sns.countplot(x='Sex', data=data, ax=ax)
ax.set_title('Gender Distribution')
st.pyplot(fig)

# Pclass vs Survived

fig, ax = plt.subplots()
