# When to use extract() and split()

When to extract() and split() - it never change the source file
1. split(): If the data present in row is all we want and seperated by a delimiter then we apply split.
    like: 30-Sep-17 --> here we want 30, Sep, 17 and '-' is delimiter on basis which we split.
        
    1. r'[\s,-]' ensures splitting by space, comma, and hyphen. expand=True makes sure results are spread across multiple columns.
    2. [...] → Matches any of these characters.
    
2. extract(): If we don't want some value in the row data then we use this.
    like: 5th Dec 2012---> here 'th' is unwanted. so if we extract on the basis of space then the result will be
        5th   Dec   2012
    we don't want it. So we have use extract() so that we can split arbitrarily as per our need.
    (\d+)(th)?\s+(\w+)\s+(\d+)
    (\d+) ------ it will extract one or more than one digit first
    (th)? -------it will extract particular format data like here th. we can define multiple value like-- (th|ed|st)?
    \s+  --------it will identify space and skip it.
    (\w+) ------- it will Captures the month name (word characters like January, Feb, etc.)
    (\W+) -------it will Captures any non-word character (anything except letters A-Z, a-z, numbers 0-9, and underscore _)
    \s+ --------another space,it will identify space and skip it.
    (\d+) ------Captures the day (one or more digits: 1, 23, etc.)
    
    5hr 25min 1sec--We have to split want 5  25  1 so there are at least one char which is unwanted so we have to extract.
    new_time = df_address["Time"].str.extract(r'(\d+)(hr)?\s+(\d+)(min)?\s+(\d+)(sec)?', expand=True)
    (\d+) ------ it will extract one or more than one digit first
    (hr)? -------it will extract particular format data like here hr. we can define multiple value like-- (th|ed|hr)?
    \s+  --------it will identify space and skip it.
    (\d+) ------Captures the day (one or more digits: 1, 23, etc.)
    (min)? -------it will extract particular format data like here min. we can define multiple value like-- (th|ed|min)?
    \s+  --------it will identify space and skip it.
    (\d+) ------Captures the day (one or more digits: 1, 23, etc.)
    (sec)? -------it will extract particular format data like here sec. we can define multiple value like-- (th|ed|sec)?

Use Case ----------------------------------              Code ----------------------------------------------Output
Basic split (default whitespace)----            df['col'].str.split()-------------------------Returns a list of words
Split by comma ---------------------------------df['col'].str.split(',')----------------------List of values split by ,
Expand into multiple columns --------------df['col'].str.split(',', expand=True) -----------DataFrame with separate columns
Limit number of splits ---------------------df['col'].str.split(',', n=1, expand=True) -------------Splits only once
Split by multiple delimiters ---------------df['col'].str.split(r'[,-]', regex=True, expand=True)----Splits by , or -


(\W+)

Symbol	Meaning
\W ----------Matches any non-word character (anything except letters A-Z, a-z, numbers 0-9, and underscore _)
+  ----------Matches one or more occurrences of the preceding pattern (\W)
() ----------Captures the matched non-word characters into a group
